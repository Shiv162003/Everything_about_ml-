```python
driver = webdriver.Chrome()
driver.get(website)

all_matches_buttons = driver.find_elements(By.XPATH, '//label[@analytics-event="All matches"]')
if all_matches_buttons:
    all_matches_buttons[0].click()
matches = driver.find_elements(By.TAG_NAME, 'tr')
for x in matches:
    print(x.text)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Set up the WebDriver with the correct path
driver = webdriver.Chrome()
# Open Google login page
driver.get("https://accounts.google.com/signin")

# Input email
email_input = driver.find_element(By.XPATH, '//*[@id="identifierId"]')
email_input.send_keys('shivanshnautiyal30@gmail.com')
email_input.send_keys(Keys.RETURN)
time.sleep(2)  # Wait for the next page to load

# Input password
password_input = driver.find_element(By.XPATH, '//*[@name="password"]')
password_input.send_keys('shiv162003')
password_input.send_keys(Keys.RETURN)
time.sleep(2)  # Wait for the next page to load

# At this point, you are logged in and can browse the internet
# Example: Open Google homepage
driver.get("https://www.google.com")
time.sleep(2)  # Wait for the page to load

# You can now interact with the page, search for something, etc.
search_box = driver.find_element(By.NAME, 'q')
search_box.send_keys('OpenAI')
search_box.send_keys(Keys.RETURN)

# Wait for a few seconds to see the results
time.sleep(5)

# Close the browser
driver.quit()
```
